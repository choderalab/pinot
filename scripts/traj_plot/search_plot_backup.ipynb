```python
from rdkit import Chem
mols = [Chem.MolFromSmiles(smiles) for smiles in moonshot_smiles]
```

```python
from rdkit.Chem import MACCSkeys
import numpy as np
fps = np.array([MACCSkeys.GenMACCSKeys(x) for x in mols])
```

```python
from sklearn.decomposition import PCA
```

```python
y = np.nan_to_num(df["f_inhibition_at_20_uM"].values, 0.0)
y = (y - np.mean(y)) / np.std(y)
```

```python
fps_cov = np.cov(fps)

plt.imshow(fps_cov)
plt.colorbar()
```

```python
fps_transpose_cov = np.cov(fps.transpose())

plt.imshow(fps_transpose_cov)
plt.colorbar()

```

```python
# @gif.frame
def plot(num_steps, traj):
    import seaborn as sns
    fig = plt.figure()
    ax = fig.add_subplot(111)

    idxs = np.argsort(y)

    sns.kdeplot(
        x[:,0],
        x[:,1],
        shade=False,
        ax=ax,
        color="black",
        alpha=0.5,
        zorder=0,
        # cmap=sns.color_palette("coolwarm", as_cmap=True),
        # hue="series"
    )

    ax.scatter(
        x[idxs][:, 0],
        x[idxs][:, 1],
        # c=y[idxs],
        c = series,
        cmap="coolwarm",
        alpha=0.8,
    )

    ax.set_xlabel("PCA0")
    ax.set_ylabel("PCA1")

    ax.set_xticks([])
    ax.set_yticks([])


    ax.plot(
        x[traj[:num_steps]][:, 0],
        x[traj[:num_steps]][:, 1],
    )

```
